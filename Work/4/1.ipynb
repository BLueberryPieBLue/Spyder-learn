```python
import requests
import urllib.request
from urllib.request import urlopen
from bs4 import BeautifulSoup 
import time
import os

R=set()
S=set()
def save(imgurl,path):
    try:
        time.sleep(1)
        header = {"User-Agent":"Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36"}
        data = requests.get(imgurl, headers = header)
        Path=path+str(time.time())+".png"
        print(Path)
        fp = open(Path,'wb')
        fp.write(data.content)
        fp.close()
    except Exception as e:
        print(e)
def select(url,select):
    try:
        R=set()
        time.sleep(1)
        req = urllib.request.Request(url)
        req.add_header("User-Agent","Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36")
        html = urllib.request.urlopen(req).read()
        soup = BeautifulSoup(html, 'html.parser')
        img = soup.select(select)
        path="./img/"
        for i in img:
            imgurl=i.get("src")
            if imgurl.startswith("//"):
                imgurl="http:"+imgurl
                print(imgurl)
                R.add(imgurl)
            if imgurl.startswith("http://") or imgurl.startswith("https://"):
                imgurl=imgurl
                print(imgurl)
                R.add(imgurl)
        for i in R:
            save(i,path)
    except Exception as e:
        print(e)
def spider(url,select):
    try:
        time.sleep(1)
        req = urllib.request.Request(url)
        req.add_header("User-Agent","Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36")
        html = urllib.request.urlopen(req).read()
        soup = BeautifulSoup(html, 'html.parser')
        a = soup.select(select)
        for i in a:
            print(i)
            if 'href' in i.attrs:
                href=i.attrs['href']
                if href.startswith("//"):
                    href="http:"+href
                    S.add(href)
                if href.startswith("http://"):
                    S.add(href)
                if href.startswith("https://"):
                    S.add(href)
    except Exception as e:
        print(e)
url = ["https://sou.autohome.com.cn/zonghe?entry=27&page={}&q=%cd%be%b9%dbL".format(str(i)) for i in range(1, 77)]+[
    "https://sou.autohome.com.cn/zonghe?q=%c6%df%d7%f9%cd%be%b9%db&page={}".format(str(i)) for i in range(1, 67)]+[
    "https://sou.autohome.com.cn/zonghe?q=%cd%be%b9%db%ba%a3%cd%e2%b0%e6&page={}".format(str(i)) for i in range(1, 33)]+[
    "https://sou.autohome.com.cn/zonghe?q=%cd%be%b9%db%ba%a3%cd%e2%b0%e6&page={}".format(str(i)) for i in range(1, 33)]+[
    "https://sou.autohome.com.cn/zonghe?q=%c8%ab%d0%c2%cd%be%b9%db%cd%bc%c6%ac&page={}".format(str(i)) for i in range(1, 17)]+[
    "https://sou.autohome.com.cn/zonghe?q=%cd%be%b9%db2017&page={}".format(str(i)) for i in range(1, 77)]+[
    "https://sou.autohome.com.cn/zonghe?q=%b4%f3%d6%da%cd%be%b9%dbl&page={}".format(str(i)) for i in range(1, 77)
    ]
for i in url:
    S=set()
    print(i)
    spider(i,"#cresult > dl > dt > a")
    select(i,"img")
    for j in S:
        print(j)
        select(j,"img")
```
